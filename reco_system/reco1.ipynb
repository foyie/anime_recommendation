```python
anime['title']=anime['title'].str.lower()
anime['title2']=anime['title2'].str.lower()
```

```python
def conv_symb(text):
    for ch in ['[',']',"'",'"',]:
        if ch in text:
            text=text.replace(ch,"")
    text_list=list(text.split(','))
    return text_list
        
```

```python
anime[anime['synopsis'].isnull()]
```

```python
anime['synopsis']=anime['synopsis'].fillna('')
anime['studio']=anime['studio'].fillna('')
```

```python
def conv_lower(s):
    if s=='':
        return list(s)
    else:
        s=s.lower()   
        s=list(s.split(' '))        
        return s
```

```python
def conv3(text):
    if text=='':
        return list(text)
    else:    
        text=text.replace(" ","")
        text=text.lower()
        text=list(text.split(" "))
        return text    
```

```python
conv3(anime['studio'][0])
```

```python
conv_lower(anime.iloc[0].synopsis)
```

```python
anime['genre']=anime['genre'].apply(conv_symb)
anime['synopsis']=anime['synopsis'].apply(conv_lower)
anime['studio']=anime['studio'].apply(conv3)
```

```python
anime.head(5)
```

```python
anime['tags']=anime['synopsis']+anime['genre']+anime['studio']
```

```python
new_df=anime[['title', 'title2', 'tags','img_url']]
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
#the top 5000 common words
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index=(new_df[new_df['title']==movie].index[0])or(new_df[new_df['title2']==movie].index[0])
    distances=similarity[movie_index]
    movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
        #print(i[0])
```

```python
recommend('jujutsu kaisen')
```
